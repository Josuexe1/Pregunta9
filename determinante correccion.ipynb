```python
from src import descomposicion_LU  

def calc_determinante(A: list[list[float]]) -> float:
    """Función que calcula el determinante usando el método de descomposición LU.

    ## Parameters
    A: Matriz cuadrada de tamaño n x n

    ## Return
    detA: Determinante de la matriz A
    """
    # Descomposición LU de la matriz A
    L, U = descomposicion_LU(A)
    
    # El determinante de A es el producto de los determinantes de L y U
    # El determinante de L es 1 porque L tiene 1s en la diagonal principal
    # El determinante de U es el producto de sus elementos diagonales
    detA = 1
    for i in range(len(U)):
        detA *= U[i][i]  # Producto de los elementos diagonales de U
    
    return detA

# Ejemplo de uso:
A1 = [
    [-4, 2, -4, -4, 1, 2, 5, 3, 5, 1],
    [1, 0, 4, 3, 0, -2, 3, 0, 1, 5],
    [5, 5, -4, 5, -4, 2, 2, 2, 4, 4],
    [-1, 3, 4, -1, -4, 0, 5, 0, 0, 5],
    [4, 1, 4, 2, 0, 0, 3, -1, 0, 2],
    [2, -2, 1, -1, -2, -3, 2, -2, 4, -1],
    [3, -2, -3, -2, -1, -3, 5, -1, 5, 0],
    [3, 4, -3, 3, -2, 2, -4, -4, 1, 5],
    [-4, 0, 3, 3, -3, -2, -2, 0, 5, -4],
    [-2, 4, 4, -2, -1, 1, 5, -1, 3, -3],
]

detA1 = calc_determinante(A1)
print(f"El determinante de A1 es: {detA1}")
```

# Ejercicio 2

```python
def calc_determinante(A: list[list[float]]) -> float:
    """Función que calcula el determinante usando el método de eliminación de Gauss.

    ## Parameters
    A: Matriz cuadrada de tamaño n x n

    ## Return
    detA: Determinante de la matriz A
    """
    n = len(A)
    detA = 1  # Inicializamos el determinante como 1

    # Copiar la matriz A para no modificar la original
    A = [row[:] for row in A]
    
    for i in range(n):
        # Buscar el pivote máximo en la columna i
        max_row = i
        for j in range(i + 1, n):
            if abs(A[j][i]) > abs(A[max_row][i]):
                max_row = j
        
        # Si el pivote es cero, el determinante es cero
        if A[max_row][i] == 0:
            return 0
        
        # Intercambiar las filas i y max_row
        A[i], A[max_row] = A[max_row], A[i]
        detA *= (-1) ** (i != max_row)  # Cambiar el signo si se intercambiaron filas
        
        # Hacer ceros debajo del pivote en la columna i
        for j in range(i + 1, n):
            if A[j][i] != 0:
                factor = A[j][i] / A[i][i]
                for k in range(i, n):
                    A[j][k] -= factor * A[i][k]
    
    # El determinante es el producto de los elementos diagonales
    for i in range(n):
        detA *= A[i][i]
    
    return detA

# Matriz A2
A2 = [
    [2, 2, 4, 5, -2, -3, 2, -2],
    [-1, -1, 3, 2, 1, 1, -4, 4],
    [2, 5, -3, -3, -2, 2, 5, 3],
    [-2, -4, 0, 1, -1, 5, -4, -1],
    [1, -2, -1, 5, 5, 2, 1, -2],
    [5, 4, 0, 3, 4, -1, -3, -2],
    [4, -4, 1, 2, 3, 3, -1, 3],
    [-2, 1, -3, 0, 5, 4, 4, -4],
]

# Calcular el determinante de A2
detA2 = calc_determinante(A2)
print(f"El determinante de A2 es: {detA2}")
```
